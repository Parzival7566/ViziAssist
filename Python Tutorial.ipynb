Let us start with the basics. What exactly is a programming language?

Put simply, programming is giving a set of instructions to a computer to execute.

If you’ve ever cooked using a recipe before, you can think of yourself as the computer and the recipe’s author as a programmer. The recipe author provides you with a set of instructions which you read and then follow. The more complex the instructions, the more complex the result!

Programming languages are the tools we use to write instructions for computers to follow. 
Computers “think” in binary — strings of 1s and 0s. Programming languages allow us to translate the 1s and 0s into something that humans can understand and write. 
A programming language is made up of a series of symbols that serves as a bridge that allow humans to translate our thoughts into instructions computers can understand.

Now, we move on to the question; what exactly is python?

Python is a general-purpose interpreted, interactive, object-oriented, and high-level programming language. It was created by Guido van Rossum during 1985- 1990. 

Python is a high-level, interpreted, interactive and object-oriented scripting language. Python is designed to be highly readable. It uses English keywords frequently where as other languages use punctuation, and it has fewer syntactical constructions than other languages.

Let us now go ahead and execute a program that takes an input from you and says something back!

```python
a=str(input("Enter your Name : "))
print("Hello ",a,"! It is nice to meet you!")
```

Here, we use the keyword input, to take the input from you and the keyword print to provide you the output. The keyword str used above is something called a datatype. 

The data stored in memory can be of many types. For example, a person's age is stored as a numeric value and his or her address is stored as alphanumeric characters. Now, lets perform a Python program to perform addition and subtraction of two numbers.

```python
a=int(input("Enter the first number : "))
b=int(input("Enter the second number : "))
addition=a+b
subtraction=a-b
print("The sum of ",a," and ",b," is ",addition)
print("The difference of ",a," and ",b," is ",subtraction)
```

As you can see, simply by creating two variables of "addition" and "subtraction", we can store the values of the sum and difference in them irrespective of whatever numbers we input.

You can try running this program again and again with different numbers and check this for yourself!

Now, lets go ahead and have fun with strings, using a method called String Slicing!

```python
str = 'Hello World!'

print str          # Prints complete string
print str[0]       # Prints first character of the string
print str[2:5]     # Prints characters starting from 3rd to 5th
print str[2:]      # Prints string starting from 3rd character
print str * 2      # Prints string two times
print str + "TEST" # Prints concatenated string
```

Once again, you are encouraged to run the above program, after changing the values of str.

Let us now move onto something called a conditional statement. 
Decision making is anticipation of conditions occurring while execution of the program and specifying actions taken according to the conditions.

Decision structures evaluate multiple expressions which produce TRUE or FALSE as outcome. You need to determine which action to take and which statements to execute if outcome is TRUE or FALSE otherwise. Following is the general form of a typical decision making structure found in most of the programming languages −

Lets go ahead and execute a conditional statement using a statement in python known as the if statement

```python
var = 100
if ( var == 100 ) : 
    print ("Value of expression is 100")
else:
    print ("Value of expression is not 100")
print "Good bye!"
```
